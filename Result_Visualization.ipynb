```python
import matplotlib.pyplot as plt

file_name = 'fold_performance_model_cnn3_cvr.json'
default_path = '/Users/linuszarse/Documents/UNI/Master-Uni Potsdam/3. Semester/Machine Learning 2/Klausurprojekt/Cross_Validation_Results/'
full_path = os.path.join(default_path, file_name)

with open(full_path, 'r') as file:
    results_cnn3 = json.load(file)

# Extract the histories
fold_histories = results_cnn3['fold_histories']

# Angenommen, fold_histories ist eine Liste von Dictionarys mit Verlaufsinformationen
# In diesem Beispiel nehmen wir an, dass jeder Eintrag ein Verlauf fr eine Faltung darstellt

# Fr jede Faltung in fold_histories
for fold_index, fold_history in enumerate(fold_histories):
    # Nehmen Sie an, dass 'loss' und 'accuracy' in jedem Verlauf vorhanden sind
    loss = fold_history['loss']
    accuracy = fold_history['accuracy']
    
    # Erstellen Sie Diagramme fr Loss und Accuracy fr jede Faltung
    plt.figure(figsize=(12, 4))
    
    # Diagramm fr Loss
    plt.subplot(1, 2, 1)
    plt.plot(loss, label='Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title(f'Faltung {fold_index + 1} Loss')
    plt.legend()
    
    # Diagramm fr Accuracy
    plt.subplot(1, 2, 2)
    plt.plot(accuracy, label='Accuracy')
```
